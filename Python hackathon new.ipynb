```python
# 3. Create a figure object called fig using plt.figure() **
# ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
 #** Plot (x,y) on that axes and set the labels.
# Note : You can plot on any x,y set of values.Create a figure object called fig using plt.figure() **



fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_ylabel('Diagnosis', color='red')
ax.set_xlabel('Patient', color='blue')

for i in range(2):
    ax.plot(np.arange(1., 0., -0.1) * 2000., np.arange(1., 0., -0.1))
    ax.set_ylabel('Diagnosis %d' % i)
    ax.set_xlabel('Patient %d' % i)
    if i == 0:
        ax.tick_params(axis='x', rotation=55)
plt.show()
```

```python
#4 Display data by splitting age in 4 quartiles and labeling the quartiles.
import pandas as pd
import numpy as np
import datetime
dfPatients = pd.read_excel("HospitalDatabase.xlsx","Patients")
#dfPatients = dfPatients['DOB'] = pd.to_datetime(dfPatients.DateOfBirth)
def from_dob_to_age(born):
    today = datetime.date.today()
    #print (today.month)
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

dfPatients['Age'] = dfPatients['DateOfBirth'].apply(lambda x: from_dob_to_age(x))
dfPatients['Age'].quantile([0.25, 0.50, 0.75, 1.0], interpolation='nearest')
```

```python
#11 Create a pie chart by taking Service= ["ICU", "Neurology", "General", "Medicine"] & ExpectedLOS=[10,5,7,9] & explode the wedge with highest LOS.

import pandas as pd
import numpy as np
import seaborn as sns

Service = ["ICU", "Neurology", "General", "Medicine"]
ExpectedLOS=[10, 5, 7, 9]
outside = (0.5, 0, 0, 0) 
# Pie Chart
#startangle to start first data rotated 90 degree counterclcokwise from x-axis
plt.pie(ExpectedLOS, labels=Service, explode=outside, autopct="%.1f", shadow=True, startangle=90) 
plt.title("Pie Chart")
plt.show()
```

```python
#13 Create a subplot on x = np.arange(0, 10, 0.1) , y = np.sin(np.pi * x) + x with 2 rows & 2 columns with object oriented approach.
import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(0, 10, 0.1)
y = np.sin(np.pi * x) + x
 
fig, ax = plt.subplots(3)

ax[0].plot(x, y)
ax[1].plot(x,y)
ax[2].plot(y,x)

ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[1].set_xlabel("x1")
ax[1].set_ylabel("y1")

ax[2].set_xlabel("x2")
ax[2].set_ylabel("y2")
```
